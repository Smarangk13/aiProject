```python
import pandas as pd
import numpy as np
```

```python
fileName = "IMDb.csv"
df = pd.read_csv(fileName)
```

```python
df.head()
```

```python
extras = ['imdb_title_id','original_title','date_published','language','description','budget',
          'votes','usa_gross_income','worlwide_gross_income','reviews_from_users','reviews_from_critics']

reviewsDf = df.drop(columns=extras)
reviewsDf = reviewsDf.dropna(how='any').reset_index()
reviewsDf.head()
```

```python
# Splits list with multiple values into a 2d list
def multiLabelSplitter(itemList):
    genre1 = []
    genre2 = []
    genre3 = []
    genreList = [genre1, genre2, genre3]
    
    for items in itemList:
        itemsX = items.split(',')
        
        l = len(itemsX)
        for i,item in enumerate(itemsX):
            item = item.strip()
            genreList[i].append(item)
        
        if l < 2:
            genreList[1].append('None')
        
        if l<3:
            genreList[2].append('None')
        
    return genreList
```

```python
genreSplit = multiLabelSplitter(reviewsDf['genre'])
# print(genreCodes)
```

```python
reviewsDf['genre 1'] = genreSplit[0]
reviewsDf['genre 2'] = genreSplit[1]
reviewsDf['genre 3'] = genreSplit[2]
```

```python
reviewsDf.head()
```

```python
reviewsDf = reviewsDf.drop(columns=['index'])
```

```python
vote = np.array(reviewsDf['avg_vote'])
vote *= 10
reviewsDf['avg_vote'] = vote
```

```python
reviewsDf.to_csv('imdb-AIClass.csv', index = False)
```
